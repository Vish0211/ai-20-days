```python
from transformers import pipeline
```

```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)
```

```python
prompt = "Artificial Intelligence is"

result = generator(
    prompt,
    max_length=50,
    num_return_sequences=1
)

print(result[0]["generated_text"])
```

```python
def assistant(question):
    prompt = f"""
    Answer the following question clearly and simply.

    Question: {question}

    Answer:
    """

    result = generator(
        prompt,
        max_length=100,
        num_return_sequences=1
    )

    return result[0]["generated_text"]

print(assistant("What is Machine Learning?"))
```

```python
print(assistant("Explain Deep Learning in 3 lines"))
```

```python
print(assistant("What is Python?"))
```

```python
print(assistant("Give 3 benefits of Artificial Intelligence"))
```

# Day 17 - Building with LLMs

## What is an LLM?
A Large Language Model (LLM) is an AI model trained on vast amounts of text data that can understand and generate human-like language.

## What is an API?
An API (Application Programming Interface) allows applications to communicate with services and exchange data.

## What is a Prompt?
A prompt is the instruction or input given to an AI model to generate a response.

## What is Prompt Engineering?
Prompt Engineering is the practice of designing effective prompts to obtain accurate and useful outputs from AI models.
