```python
from transformers import pipeline

sentiment = pipeline("sentiment-analysis")

print(sentiment("I love learning AI"))
print(sentiment("I am tired today"))
```

```python
from transformers import pipeline

ner = pipeline("ner", aggregation_strategy="simple")

text = "Elon Musk founded Tesla in the United States."

result = ner(text)

for entity in result:
    print(entity)
```

# Day 16 - Transformers with Hugging Face

## What is a Transformer?
A Transformer is a deep learning model that understands the context and meaning of text using an attention mechanism.

## What is Hugging Face?
Hugging Face is a platform and Python library that provides pretrained AI models for NLP, computer vision, and other AI tasks.

## What is a Pipeline?
A pipeline is a ready-to-use AI task that allows us to perform operations such as sentiment analysis, named entity recognition, and text generation with just a few lines of code.

## Tasks Performed
1. Sentiment Analysis
2. Named Entity Recognition (NER)

## Key Learning
Pretrained transformer models can perform complex NLP tasks without requiring us to train models from scratch.
