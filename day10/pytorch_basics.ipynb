```python
import torch

# fake data
x = torch.randn(100, 1)
y = 2*x + 1 + torch.randn(100, 1) * 0.1
```

```python
import torch

# initialize parameters randomly
w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

print("Initial weight:", w)
print("Initial bias:", b)
```

```python
learning_rate = 0.01

for epoch in range(100):

    # prediction
    y_pred = w*x + b

    # loss (error)
    loss = ((y_pred - y)**2).mean()

    # backward (calculate gradients)
    loss.backward()

    # update weights
    with torch.no_grad():
        w -= learning_rate * w.grad
        b -= learning_rate * b.grad

        # reset gradients
        w.grad.zero_()
        b.grad.zero_()

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item()}")
```

```python
print("Final weight (w):", w.item())
print("Final bias (b):", b.item())
```

```python
x_test = torch.tensor([[1.0], [2.0], [3.0]])

y_pred = w*x_test + b

print("Predictions:")
print(y_pred)
```
